# Solution 1

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import sys
```

```python
a = int(input("Enter value of a: "))

b = int(input("Enter value of b: "))
```

```python
mult_fact = 1000000
ep = sys.float_info.epsilon
```

```python
def err_sum(n_tot):
    
    e1 = []
    e2 = []
    e3 = []
    o_index = []
    
    for n in range(1, n_tot+1):
    
        lst = np.array([(random.uniform(a, b) + (mult_fact*ep)) for i in range(n)])
    
        #Unsorted Sum
        s1 = sum(lst)
    
        #Sorted Sum
        s2 = sum(np.sort(lst))
    
        #Compensated Sum
        s3 = 0.0
        c = 0.0
        for i in range(0, n):
            y = lst[i] - c
            t = s3 + y
            z = t - s3
            c = z - y
            s3 = t
        
        e1.append(np.abs(s1 - s3))
        e2.append(np.abs(s1 - s3))
        e3.append(np.abs(s2 - s3))
        o_index.append(n)
    
    error_pd = pd.DataFrame(list(zip(e1, e2, e3)), index = o_index, 
                            columns = ['Unsorted-Sorted', 'Unsorted-Compensated', 'Sorted-Compensated'])
    
    return error_pd
```

```python
n_max = 100
err = err_sum(n_max)
```

```python
plt.plot(err['Unsorted-Sorted'])
plt.xlabel("Number of Elements in Array")
plt.ylabel("Error Value")
plt.title("Unsorted Sum - Sorted Sum")
plt.show()
```

```python
plt.plot(err['Unsorted-Compensated'])
plt.xlabel("Number of Elements in Array")
plt.ylabel("Error Value")
plt.title("Unsorted Sum - Compensated Sum")
plt.show()
```

```python
plt.plot(err['Sorted-Compensated'])
plt.xlabel("Number of Elements in Array")
plt.ylabel("Error Value")
plt.title("Sorted Sum - Compensated Sum")
plt.show()
```

```python
import math
```

```python
c = float(input("Enter value of c: "))

i = float(input("Enter value of i: "))
```

```python
n = 365
x = i/365
```

```python
ci_vanilla = c*(math.e**(n*math.log(1+x))-1)/x
ci_vanilla
```

```python
ci_misc =  c*(math.e**(n*((x*math.log(1+x))/((1+x)-1)))-1)/x
ci_misc
```

```python
log_mac = 0.0                  

for i in range(1, 366):
    log_mac += (-1)**(i+1)*(x**i)/i
    
ci_maclaurin = c*(math.e**(n*log_mac)-1)/x    
ci_maclaurin
```
