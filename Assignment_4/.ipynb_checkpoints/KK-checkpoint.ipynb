```python
import os 
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import statsmodels.api as sm
from sklearn.metrics import r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.linear_model import LinearRegression, Lasso, Ridge, lasso_path, lars_path, enet_path
%run script.py
```

## Data Cleanup

### 1.1 Load data and pre-processing of the data

```python
data = pd.read_csv('./data/train.csv')
data.head()
```

```python
y = data['critical_temp']
X = data[data.columns[['wtd' in column for column in data.columns]]]
```

```python
# data.isnull().sum()
# data.info()
# data.describe()
```

```python
cor_mat = X.corr().abs()
plt.figure(figsize=(25,9))
sns.heatmap(X.corr(), annot = True)
plt.title("Correlation matrix")
plt.show()
```

### 1.2 Filter by correlation coefficient $\leq$ 0.8

```python
# upper_tri = X.corr().where(np.triu(np.ones(cor_mat.shape),k=1).astype(np.bool))
# to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.8)]
# # X1 = X.drop(to_drop , axis = 1)
# to_drop
```

```python
l = (cor_mat>0.8).sum() == 1
X = X[X.columns[l]]
```

### 1.3 Print VIF's

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]

vif_data = vif_data.sort_values(by = ['VIF'])
vif_data
```

```python
# top_19 = vif_data.feature[:19]
# sum([x in to_drop for x in list(top_19)])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = False)
print(f'X_train: {X_train.shape}\ny_train: {y_train.shape}')
print(f'X_test: {X_test.shape}\ny_test: {y_test.shape}')
```

## Fit OLS on important predictors

### 2.1 OLS Model Fitting

```python
mlr_model = LinearRegression().fit(X_train, y_train)
mlr_model.coef_
```

```python
print(f'Model Score: {np.round(mlr_model.score(X_test, y_test)*100, 2)}%')
```

## Fitting Lasso Regression

```python
lasso_model = Lasso(alpha=0.9)
lasso_model.fit(X_train, y_train)
lasso_model.score(X_test, y_test)
```

```python
print(f'Model Score: {np.round(lasso_model.score(X_test, y_test)*100, 2)}%')
```

###  3.1 Compare the coefficients of selected variables with OLS model

```python
plt.figure(figsize = (16,9))
xx = np.linspace(1,12,12)
plt.plot(xx, mlr_model.coef_, label = 'OLS', color = 'red')
plt.plot(xx, lasso_model.coef_, label = 'Lasso')
plt.xlabel(''); plt.ylabel('Coefficient'); plt.title('')
plt.legend(loc = 'best')
plt.show()
```

### 3.2 Plot LASSO path

```python
alpha, coefs, _ = lasso_path(np.array(X_train), np.array(y_train))
coefs = pd.DataFrame(coefs.T, columns = X_train.columns)
xx = np.sum(np.abs(np.array(coefs)), axis=1)
xx /= xx[-1]
plt.figure(figsize = (16,9))
plt.plot(xx, coefs)
plt.legend(X_train.columns, bbox_to_anchor = (1,1))
plt.title("LASSO Path")
plt.axis("tight")
plt.show()
```

```python
clf = Lasso()
coefs = []
alphas = np.logspace(-6, 6, 100)
for i in alphas:
    clf.set_params(alpha = i)
    clf.fit(X_train, y_train)
    coefs.append(clf.coef_)
```

```python
plt.figure(figsize = (16,9))
ax = plt.gca()
ax.set_xscale("log")
ax.plot(alphas, coefs)
plt.xlabel("alpha")
plt.ylabel("weights")
plt.title("Lasso coefficients as a function of the regularization")
plt.axis("tight")
plt.show()
```

### 3.3 Lasso Lars Path

```python
_, _, coefs = lars_path(np.array(X_train), np.array(y_train), method = "lasso", verbose=True)
coefs = pd.DataFrame(coefs.T, columns = X_train.columns)
xx = np.sum(np.abs(np.array(coefs)), axis=1)
xx /= xx[-1]
plt.figure(figsize = (16,9))
plt.plot(xx, coefs)
plt.legend(X_train.columns, bbox_to_anchor = (1,1))
ymin, ymax = plt.ylim()
plt.vlines(xx, ymin, ymax, linestyle="dashed")
plt.xlabel("|coef| / max|coef|")
plt.ylabel("Coefficients")
plt.title("LASSO Path")
plt.axis("tight")
plt.show()
```

# Fitting Ridge Regression 

```python
ridge_model = Ridge(alpha = 100)
ridge_model.fit(X_train, y_train)
ridge_model.score(X_test, y_test)
```

```python
print(f'Model Score: {np.round(ridge_model.score(X_test, y_test)*100, 2)}%')
```

<h4 style = 'color: blue'> 4.2 Plot Ridge path

```python
clf = Ridge()
coefs = []
alphas = np.logspace(-6, 6, 100)
for i in alphas:
    clf.set_params(alpha = i)
    clf.fit(X_train, y_train)
    coefs.append(clf.coef_)
```

```python
plt.figure(figsize = (16,9))
ax = plt.gca()
ax.set_xscale("log")
ax.plot(alphas, coefs)
plt.xlabel("alpha")
plt.ylabel("weights")
plt.title("Ridge coefficients as a function of the regularization")
plt.axis("tight")
plt.show()
```
